```python
'''
Logistic regression is used to solve classification problem. They mostly used categorical values
There are 2 types of classification
1)Binary Classification(yes or no, hor or cold)etc
2)Multiclass Classification this is where by you have more that two options to choose from eg voting(Democratic, republican, independent)
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('insurance_data.csv')
df.head()
```

```python
plt.scatter(df.age, df.bought_insurance, color = 'blue', marker ='*')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['age']],df.bought_insurance, test_size = 0.1)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
#This is used for training
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
model.score(  x_test, y_test)
```

```python
model.predict_proba(x_test)
```
