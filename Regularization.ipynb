```python
'''
How to reduce overfitting: Shrinking the parameters
We shrink the parameters by penalizing the coeffient more specifically penalizing the theta in which is a square in L2(ridge regression) and absolute theta in L1(lasso regression)
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#Supress warnings for clean collab
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('Melbourne_housing_FULL.csv')
dataset.head()
```

```python
dataset.nunique()
```

```python
dataset.shape
```

```python
useful_columns = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 'Distance', 'CouncilArea', 'Bedroom2',
                  'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']

dataset = dataset[useful_columns]
dataset.head()
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
dataset.isna().sum()
```

```python
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset.isna().sum()
```

```python
dataset.dropna(inplace = True)
dataset.isna().sum()
```

```python
dataset = pd.get_dummies(dataset, drop_first = True).astype(int)
dataset.head()
```

```python
X = dataset.drop('Price', axis = 1)
Y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.score(x_train, y_train)
```

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha = 50, max_iter = 100, tol = 0.1)
lasso_reg.fit(x_train, y_train)
```

```python
lasso_reg.score(x_test, y_test)
```

```python
lasso_reg.score(x_train, y_train)
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha = 50, max_iter = 100, tol = 0.1)
ridge_reg.fit(x_train, y_train)
```

```python
ridge_reg.score(x_test, y_test)
```
