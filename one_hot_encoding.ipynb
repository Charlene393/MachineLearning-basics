```python
import pandas as pd
df = pd.read_csv("homeprices.csv")
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
!pip install word2number
```

```python
merged = pd.concat([df, dummies], axis = 1)
merged[['monroe township', 'robinsville', 'west windsor']] = merged[['monroe township', 'robinsville', 'west windsor']].astype(int)
merged
```

```python
final = merged.drop(['town', 'west windsor'], axis = 'columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#Because it is a dependent variable
X = final.drop('price', axis= 'columns')
X
```

```python
y = final.price
y
```

```python
model.fit(X, y)
```

```python
#prediction for robinsville
model.predict([[2800, 0,1]])
```

```python
#prediction for west windisor
model.predict([[3000,0,0]])
```

```python
#Checking the accuracy of the model
model.score(X,y)
```
